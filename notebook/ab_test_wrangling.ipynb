## __Check and Clean Data__

```python
def group_and_landing_page_integrity_check(df: pd.DataFrame):
    df = (df
     .assign(group_numeric=lambda df_: df_["group"].apply(lambda row: 0 if row == "control" else 1))
     .assign(landing_page_numeric=lambda df_: df_["landing_page"].apply(lambda row: 0 if row == "old_page" else 1))
    )
    try:
        assert df["group_numeric"].to_list() == df["landing_page_numeric"].to_list()
        print("KOLOM 'GROUP' DAN 'LANDING_PAGE' SESUAI.")
    except AssertionError:
        print("KOLOM 'GROUP' DAN 'LANDING_PAGE' TIDAK SESUAI.")

def clean_ab_test_data(df: pd.DataFrame) -> pd.DataFrame:
    return (df
     .assign(timestamp=lambda df_: pd.to_datetime(df_["timestamp"]))
     .assign(landing_page=lambda df_: df_["group"].apply(lambda x: "old_page" if x == "control" else "new_page"))
     .drop_duplicates(subset=["user_id"], ignore_index=True)
     .assign(date=lambda df_: df_["timestamp"].dt.date)
    )

def create_conversion_rate_trend(df: pd.DataFrame) -> pd.DataFrame:
    df = ab.groupby(["date", "group"]).agg({"converted":["sum", "count"]}).reset_index()
    df.columns = df.columns.droplevel(0)
    df.columns = ["date", "group", "converted", "visit"]
    return (df.assign(converted_cumulative=lambda df_: df_.groupby("group")["converted"].cumsum(),
                   visit_cumulative=lambda df_: df_.groupby("group")["visit"].cumsum(),
                   conversion_rate=lambda df_: df_["converted_cumulative"]/df_["visit_cumulative"])
            .pivot_table(index="date", columns="group", values="conversion_rate").reset_index()
           )

def create_power_trend(df_ori: pd.DataFrame, alpha=0.05, test_split=0.5, relative_effect=0.15) -> pd.DataFrame:
    df = df_ori.groupby("date").agg({"converted":["sum", "count"]}).reset_index()
    df.columns = df.columns.droplevel(0)
    df.rename(columns={"sum":"converted", "count":"visit"}, inplace=True)
    return (df.assign(converted_cumulative=lambda df_: df_["converted"].cumsum(),
                   visit_cumulative=lambda df_: df_["visit"].cumsum(),
                   conversion_rate=lambda df_: df_["converted_cumulative"]/df_["visit_cumulative"],
                   power=lambda df_: df_.apply(lambda df_: power_proportions_2indep(diff=df_["conversion_rate"]*relative_effect,
                                                                                  prop2=df_["conversion_rate"],
                                                                                  nobs1=df_["visit_cumulative"]*test_split,
                                                                                  ratio=(1-test_split)/test_split,
                                                                                  alpha=alpha,
                                                                                  alternative="two-sided",
                                                                                  return_results=False), axis=1))
           )

def ab_test_overall(df: pd.DataFrame):
    ab_groupby_test = df.groupby("group").agg({"converted":["sum", "count"]}).reset_index()
    ab_groupby_test.columns = ab_groupby_test.columns.droplevel(0)
    ab_groupby_test.rename(columns={"sum":"converted", "count":"visit"}, inplace=True)
    print (test_proportions_2indep(count1=ab_groupby_test["converted"][1],
                           nobs1=ab_groupby_test["visit"][1],
                           count2=ab_groupby_test["converted"][0],
                           nobs2=ab_groupby_test["visit"][0],
                           alternative="two-sided"))
    return ab_groupby_test

def daily_ab_test(df: pd.DataFrame) -> pd.DataFrame:
    ab_test_daily = (df
     .groupby(["date", "group"])
     .agg({"converted":["sum", "count"]})
     .reset_index()
    )
    ab_test_daily.columns = ab_test_daily.columns.droplevel(0)
    ab_test_daily.columns = ["date", "group", "converted", "visit"]
    ab_test_daily = ab_test_daily.assign(converted_cumulative=lambda df_: df_.groupby("group")["converted"].cumsum(),
                                         visit_cumulative=lambda df_: df_.groupby("group")["visit"].cumsum())
    daily_ab_test_result = pd.DataFrame()
    for date in ab_test_daily["date"].unique():
        df_new = pd.DataFrame()
        df_new["date"] = pd.Series(date)
        df_filtered = ab_test_daily.query("date == @date").reset_index()
        df_new["pvalue"] = pd.Series(test_proportions_2indep(count1=df_filtered["converted_cumulative"][1],
                                                                nobs1=df_filtered["visit_cumulative"][1],
                                                                count2=df_filtered["converted_cumulative"][0],
                                                                nobs2=df_filtered["visit_cumulative"][0],
                                                                alternative="two-sided").pvalue)
        daily_ab_test_result=pd.concat([daily_ab_test_result, df_new], ignore_index=True)
    return daily_ab_test_result
```
